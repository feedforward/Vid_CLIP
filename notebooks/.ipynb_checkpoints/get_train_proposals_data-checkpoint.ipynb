```python
# Import statements
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
import PIL.Image as Image
import cv2
import pandas as pd
import random 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load inputs
act_map = {}
with open("../../../act_idx_to_name_new.txt") as f:
	for line in f:
		(key, val) = line.split()
		act_map[int(key)] = val
```

```python
train_val_pd_path = '/data2/puppala/projects/text_embedding/meta.csv'
# feat_dirs = {'other': '/ssd1/seunghwc/train_feat'}
# feat_dir = feat_dirs['other']
# feat_files = sorted(os.listdir(feat_dir), key = lambda x: (len (x), x))

pd_frame_raw = pd.read_csv(train_val_pd_path)
pd_frame = pd_frame_raw.drop_duplicates()
activity_list = [x for x in pd_frame['Activity_Id'].tolist() if x]
activity_count = len(activity_list)
fig = plt.figure(figsize=(20, 10))
#hist = np.histogram(pd_frame['Activity_Id'].tolist(), bins=range(38))
arr= plt.hist(activity_list, bins=range(38))
plt.xticks(range(38), list(act_map.values()), rotation='vertical')

for i in range(37):
    plt.text(arr[1][i],arr[0][i],str(int(arr[0][i])))
```

```python
num_samples = pd_frame.shape[0]
indices = list(pd_frame.index)
data = {}
for i in indices:
	if(pd_frame['Activity_Id'][i]):
		data[i] = {'Activity_Id': pd_frame['Activity_Id'][i],
				   'Video_Path': pd_frame['Video_Path'][i],
				   'Start_Frame': pd_frame['Start_Frame'][i],
				   'End_Frame': pd_frame['End_Frame'][i],
				   'BBox': [pd_frame['x0'][i], pd_frame['y0'][i], pd_frame['x1'][i], pd_frame['y1'][i]]
				  }
```

```python
np.save('train_data.npy', data)
```
